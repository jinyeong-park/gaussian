```python
from matplotlib import pyplot as mp
import numpy as np

def gaussian(x,mu, sigma):
    return np.exp(-np.power(x-mu, 2.0) / (2* np.power(sigma,2.)))

input_arr = [(0,1), (0,2), (1,0.5), (1,0.75)]

for mu, sigma in input_arr:
    mp.plot(np.linspace(-5,5, num=100),gaussian(np.linspace(-5,5,num=100), mu, sigma), label="mu={} and sigma={}".format(mu, sigma))
    mp.legend(loc="upper left", ncol=1, title="y=np.exp(-np.power(x-mu, 2.0) / (2*np.power(sigma,2.)))")
    mp.show()
```
